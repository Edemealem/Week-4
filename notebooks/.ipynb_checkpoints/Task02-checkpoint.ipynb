### Step 1: Load the Data

```python
import pandas as pd
# Load data
file_path = 'C:/Users/user/Desktop/data_dirashe.csv'
data = pd.read_csv(file_path)
```

```python
data.head()
```

### Step 2: Data Preparation

```python
#2.1: Create Engagement Metrics

# Convert relevant columns to numeric if they're not already
data['Dur. (ms)'] = pd.to_numeric(data['Dur. (ms)'], errors='coerce')
data['HTTP DL (Bytes)'] = pd.to_numeric(data['HTTP DL (Bytes)'], errors='coerce')
data['HTTP UL (Bytes)'] = pd.to_numeric(data['HTTP UL (Bytes)'], errors='coerce')

# Compute session frequency, total duration, and total traffic per user (MSISDN)
engagement_metrics = data.groupby('MSISDN/Number').agg(
    session_frequency=('Bearer Id', 'count'),
    duration=('Dur. (ms)', 'sum'),
    total_dl_traffic=('HTTP DL (Bytes)', 'sum'),
    total_ul_traffic=('HTTP UL (Bytes)', 'sum')
).reset_index()

# Calculate total traffic by summing download and upload traffic
engagement_metrics['total_traffic'] = engagement_metrics['total_dl_traffic'] + engagement_metrics['total_ul_traffic']

# Drop the individual download and upload columns if needed
engagement_metrics.drop(columns=['total_dl_traffic', 'total_ul_traffic'], inplace=True)

engagement_metrics.head()
```

### Step 3: Aggregate Metrics and Report Top Customers

```python
#3.1: Top 10 Customers per Engagement Metric
top_customers_frequency = engagement_metrics.nlargest(10, 'session_frequency')
top_customers_duration = engagement_metrics.nlargest(10, 'duration')
top_customers_traffic = engagement_metrics.nlargest(10, 'total_traffic')

top_customers_frequency, top_customers_duration, top_customers_traffic
```

### Step 4: Normalize Metrics and K-Means Clustering

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Normalize the metrics
scaler = StandardScaler()
normalized_metrics = scaler.fit_transform(engagement_metrics[['session_frequency', 'duration', 'total_traffic']])

# K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
engagement_metrics['cluster'] = kmeans.fit_predict(normalized_metrics)

# Compute statistics for each cluster
cluster_stats = engagement_metrics.groupby('cluster').agg(
    min_frequency=('session_frequency', 'min'),
    max_frequency=('session_frequency', 'max'),
    avg_frequency=('session_frequency', 'mean'),
    total_frequency=('session_frequency', 'sum'),
    min_duration=('duration', 'min'),
    max_duration=('duration', 'max'),
    avg_duration=('duration', 'mean'),
    total_duration=('duration', 'sum'),
    min_traffic=('total_traffic', 'min'),
    max_traffic=('total_traffic', 'max'),
    avg_traffic=('total_traffic', 'mean'),
    total_traffic=('total_traffic', 'sum')
).reset_index()

cluster_stats
```

### Step 5: Visualize Cluster Statistics

```python
import seaborn as sns

# Plotting average metrics per cluster
cluster_stats_melted = cluster_stats.melt(id_vars='cluster', value_vars=['avg_frequency', 'avg_duration', 'avg_traffic'])

plt.figure(figsize=(12, 6))
sns.barplot(x='variable', y='value', hue='cluster', data=cluster_stats_melted)
plt.title('Average Engagement Metrics by Cluster')
plt.ylabel('Average Value')
plt.xlabel('Metrics')
plt.show()
```

### Step 6: Engagement per Application

```python
# Aggregate user total traffic per application
app_metrics = data.groupby('Last Location Name').agg(
    total_dl_traffic=('HTTP DL (Bytes)', 'sum'),
    total_ul_traffic=('HTTP UL (Bytes)', 'sum')
).reset_index()

# Calculate total traffic
app_metrics['total_app_traffic'] = app_metrics['total_dl_traffic'] + app_metrics['total_ul_traffic']

# Derive top 10 engaged users per application
top_apps = app_metrics.nlargest(10, 'total_app_traffic')

top_apps
```

### Step 7: Plot Top 3 Most Used Applications

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='total_app_traffic', y='Last Location Name', data=top_apps)
plt.title('Top 3 Most Used Applications')
plt.xlabel('Total Traffic (Bytes)')
plt.ylabel('Application')
plt.show()
```

### Step 8: Determine Optimal Value of k Using Elbow Method

```python
inertia = []
K = range(1, 11)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(normalized_metrics)
    inertia.append(kmeans.inertia_)

# Plot the elbow method
plt.figure(figsize=(10, 6))
plt.plot(K, inertia, 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()
```

### Step 9: Interpretation of Findings
1. User Engagement Metrics Overview

2. Clustering Results
After applying K-means clustering to the normalized engagement metrics, we classified users into three distinct engagement groups. Here's a breakdown of the findings:
* Cluster Characteristics:
   - Cluster 0: Represents low engagement users. These users have minimal session frequency, shorter session durations, and lower total traffic.
   - Cluster 1: Includes moderate engagement users. They show a balanced session frequency and duration, indicating occasional use of applications.
   - Cluster 2: Captures high engagement users. This group exhibits the highest session frequency, longest durations, and significantly higher total traffic, suggesting they are heavily invested in the applications
   
3. Visual Insights
The bar plots visualizing average metrics by cluster effectively illustrate the differences in engagement levels.
* Key Observations:

   - Moderate engagement users (Cluster 1) may require targeted marketing and resource allocation to convert them into high engagement users.\
   - Low engagement users (Cluster 0) present an opportunity for improvement; understanding their barriers to engagement could enhance their experience.
